```python
https://api.bdnb.io/v1/bdnb/donnees/batiment_groupe_complet?code_commune_insee=eq.13204
```

```python
import requests
import geopandas as gpd
import osm2geojson
import json

r =  requests.get('https://api.bdnb.io/v1/bdnb/donnees/batiment_groupe_complet',
                  params= {'code_commune_insee': 'eq.13204',
                          'limit' : 10 },
                  headers = {"Accept": "application/geo+json"}
                 )
djson=r.json()
gdf = gpd.GeoDataFrame.from_features(djson)
gdf.set_crs(epsg=2154, inplace=True)
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

fig, ax = plt.subplots(figsize=(8, 8), layout='constrained')

gdf.plot(ax=ax, color='blue',marker='o',markersize=10)

legend_elements = [
    Patch(facecolor='blue', label='Buildings BDNB')
]
ax.set_title('Buildings BDNB - Marseille 4ème arrondissement')
ax.legend(handles=legend_elements, loc='lower left')

plt.show()
```

## API BDNB v2 GENOVE

https://api-portail.bdnb.io/catalog/api/04dce6f0-cb9d-3e61-9240-af6c89d76a71/doc?page=23dce4b2-211c-30d3-a681-46313c1bfd11

```python
import osmnx as ox
import geopandas as gpd
place_name="Marseille 4e arrondissement"
area = ox.geocode_to_gdf(place_name)
area.set_crs(2154 , allow_override=True)

xmin=area.bounds.get(["minx"]).iloc[0].mean()
ymin=area.bounds.get(["miny"]).iloc[0].mean()
xmax=area.bounds.get(["maxx"]).iloc[0].mean()
ymax=area.bounds.get(["maxy"]).iloc[0].mean()
area
```

```python
import requests
r = requests.get('https://api.bdnb.io/v2/gorenove/buildings/bbox', 
                     params={'xmin': xmin,
                             'ymin': ymin,
                             'xmax': xmax,
                             'ymax': ymax,
                             'limit':5 ,
                             'hauteur': 'gt.10' ,
                             'select': 'bnb_id, hauteur, adresse_postal, l_ban_id, geombui'
                            },
                  headers= {'Accept': 'application/geo+json',
                           })
r.json()
```
