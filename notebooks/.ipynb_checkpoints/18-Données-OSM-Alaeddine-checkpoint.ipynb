## Overpass

Requête overpass : https://overpass-turbo.eu/s/1OXM

```
[out:json][timeout:25];
(
  node(user:"Datagistips");
  way(user:"Datagistips");
  relation(user:"Datagistips");
);
out meta;
```

Attention, sensible à la casse : Datagistips différent de datagistips

## Lecture de la réponse

````python
r.json()['elements']
````

````python
j = r.json()['elements']
````

````python
len(j)
````

````python
ids = list()
lons = list()
lats = list()
timestamps = list()
versions = list()
tagss = list()

for elt in j:
    id = elt['id']
    if 'lon' in elt.keys():
        lon = elt['lon']
        lat = elt['lat']
    else:
        lon = None
        lat = None
    timestamp = elt['timestamp']
    version = elt['version']
    if 'tags' in elt.keys() :
        tags = elt['tags']
    else:
        tags = None
    
    ids.append(id)
    lons.append(lon)
    lats.append(lat)
    timestamps.append(timestamp)
    versions.append(version)
    tagss.append(tags)
````

````python
import pandas as pd
````

````python
df = pd.DataFrame(data = {'id' : ids, 'lon' : lons, 'lat' : lats, 'timestamp' : timestamps, 'version' : versions, 'tags' : tagss})
````

````python
import geopandas as gpd
````

````python
pts = gpd.points_from_xy(df.lon, df.lat)
pts
````

````python
gdf = gpd.GeoDataFrame(
    df, geometry=pts, crs="EPSG:4326"
)
````

````python
gdf
````

````python
gdf.to_file('OSM-Aladdin.gpkg', driver = 'GPKG')
````

## Fonction

````python
def get_osm_from_user(user): 
    user = 'Alaeddinejerad'
    overpass_url = "http://overpass-api.de/api/interpreter"
    query = (f'[out:json][timeout:25];(node(user:"%s");way(user:"%s");relation(user:"%s"););out meta;')%(user, user, user)
    r = requests.get(overpass_url, params={'data': query})

    ids = list()
    lons = list()
    lats = list()
    timestamps = list()
    versions = list()
    tagss = list()

    for elt in j:
        id = elt['id']
        if 'lon' in elt.keys():
            lon = elt['lon']
            lat = elt['lat']
        else:
            lon = None
            lat = None
        timestamp = elt['timestamp']
        version = elt['version']
        if 'tags' in elt.keys() :
            tags = elt['tags']
        else:
            tags = None

        ids.append(id)
        lons.append(lon)
        lats.append(lat)
        timestamps.append(timestamp)
        versions.append(version)
        tagss.append(tags)

    df = pd.DataFrame(data = {'id' : ids, 'lon' : lons, 'lat' : lats, 'timestamp' : timestamps, 'version' : versions, 'tags' : tagss})

    gdf = gpd.GeoDataFrame(
        df, geometry=pts, crs="EPSG:4326"
    )
    
    return(gdf)

get_osm_from_user("Alaeddinejerad")
````
